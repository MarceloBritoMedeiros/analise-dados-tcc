```python
import requests

def get_place_id(latitude, longitude, api_key):
    url = f"https://maps.googleapis.com/maps/api/geocode/json?latlng={latitude},{longitude}&key={api_key}"
    response = requests.get(url)
    data = response.json()
    
    if 'results' in data and len(data['results']) > 0:
        place_id = data['results'][0]['place_id']
        return place_id
    else:
        return None

# Substitua 'YOUR_API_KEY' pelo seu próprio token de API do Google
latitude= -23.6244547
longitude= -46.6379525
api_key = 'AIzaSyALXezoywqm1Vfc4KCU5YEnCQOSGSvSPPM'

place_id = get_place_id(latitude, longitude, api_key)
print("Place ID:", place_id)
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', 30)
```

```python
dados_pd = pd.read_json("lotacoes_metro.json")
```

```python
dados_pd.head()
```

```python
def extrair_lat(texto):
    return texto["lat"] 

def extrair_lng(texto):
    return texto["lng"] 
```

```python
dados_pd["Latitude"] = dados_pd["coordinates"].apply(extrair_lat)
dados_pd["Longitude"] = dados_pd["coordinates"].apply(extrair_lng)
```

```python
dc = 1000
dados_pd["latitude_abr"] = dados_pd["Latitude"].apply(lambda x: float(int(x*dc))/dc)
dados_pd["longitude_abr"] = dados_pd["Longitude"].apply(lambda x: float(int(x*dc))/dc)
```

```python
import json

# Abrir o arquivo JSON e carregá-lo em um dicionário
with open('trajeto_trabalho.json', 'r') as arquivo:
    trajeto = json.load(arquivo)
```

```python
steps = trajeto["routes"][0]["legs"][0]["steps"]
```

```python
train_steps = []
for i in steps:
    try:
        o = i["transit_details"]
        if o["line"]["vehicle"]["type"] in ("SUBWAY", "HEAVY_RAIL"):
            train_steps.append(o["departure_stop"]["location"])
    except:
        pass
```

```python
train_steps
```

```python
import requests 
def geolocation(lat, lng):
    return requests.get(f"https://maps.googleapis.com/maps/api/geocode/json?key=AIzaSyALXezoywqm1Vfc4KCU5YEnCQOSGSvSPPM&latlng={lat}, {lng}").json()
```

```python
x = geolocation(train_steps[0]['lat'], train_steps[0]['lng'])
```

```python
dc = 1000
cont = 0

for i in x["results"]:
    lat1 = float(int(i["geometry"]["location"]["lat"]*dc))/dc
    lng1 = float(int(i["geometry"]["location"]["lng"]*dc))/dc

    lat2 = float(int(i["geometry"]["viewport"]["northeast"]["lat"]*dc))/dc
    lng2 = float(int(i["geometry"]["viewport"]["northeast"]["lng"]*dc))/dc

    lat3 = float(int(i["geometry"]["viewport"]["southwest"]["lat"]*dc))/dc
    lng3 = float(int(i["geometry"]["viewport"]["southwest"]["lng"]*dc))/dc    

    for lat in [lat1, lat2, lat3]:        
        for lng in [lng1, lng2, lng3]:
            if cont == 0:
                #print(f"{lat}, {lng}")
                df = dados_pd[(dados_pd["latitude_abr"]==lat)&(dados_pd["longitude_abr"]==lng)]
                if df.empty==False:
                    lotacao = df["populartimes"]
                    cont = 1    
```

```python
def converte(paulista):
    paulista_dict = {}
    for i in paulista:
        paulista_dict[i["name"]] = i["data"]
    return pd.DataFrame(paulista_dict)
```
